## Introduction to python functions

Functions allow you to reuse bits of code. You
can think of them as black boxes.

```python
def cent_to_kelv(temperature):
    temperature = temperature + 273.15
    return temperature
```

```python
t = 0.0
t_in_K = cent_to_kelv(t)
print("Temperature in C:", t)
print("Temperature in K:", t_in_K)
```

```python
def fahr_to_cent(temp):
    temp = (temp - 32) * (5/9)
    return temp
```

```python
print("-40 F is ", fahr_to_cent(-40), "C")
```

```python
print("-20 F is", fahr_to_cent(-20), "C")
```

```python
def fahr_to_kelv(temperature):
    temperature = fahr_to_cent(temperature)
    temperature = cent_to_kelv(temperature)
    return(temperature)
```

```python
t = 32
fahr_to_kelv(t)
```

```python
def fahr_to_kelv(temper):
    temper=cent_to_kelv(fahr_to_cent(temper))
    return temper
```

```python
def plot_data_from_file(filename):
    data = numpy.loadtxt(fname=filename, delimiter=',')
    fig = matplotlib.pyplot.figure(figsize=(10.0, 3.0))
    
    axes1 = fig.add_subplot(1, 3, 1)
    axes2 = fig.add_subplot(1, 3, 2)
    axes3 = fig.add_subplot(1, 3, 3)
    
    axes1.set_title (filename)
    axes1.set_ylabel('average')
    axes1.set_xlabel('timestep')
    #some code to plot the mean across axis 0
    axes1.plot(data.mean(axis=0))
    
    axes2.set_ylabel('max')
    axes2.set_xlabel('timestep')
    #some code to plot the max across axis 0
    axes2.plot(data.max(axis=0))
    
    axes3.set_ylabel('min')
    axes3.set_xlabel('timestep')
    #some code to plot the min across axis 0   
    axes3.plot(data.min(axis=0))
    
    fig.tight_layout()
    matplotlib.pyplot.show()
```

```python
def validate_data_from_file(f):
    """Read a csv file `f` and validate the data
    
       it has been observed that some of our data is
       rubbish. The maximum over all stations forms a
       triangle, or the minimum is zero (for all time).
       This checks the data in `f` and prints some 
       data on these errors.
    """
    data = numpy.loadtxt(fname= f, delimiter = ',')
    if data.max(axis=0)[0]==0 and data.max(axis=0)[20]==20:
        print(f,'Suspicious maxima')
    elif data.min(axis=0).sum() == 0:
        print(f,'Minima add up to zero')
    else:
        print(f,'Seems OK')
```

```python
%matplotlib inline
import matplotlib.pyplot
import numpy

import glob

for file in glob.glob('data/we*.csv'):
    validate_data_from_file(file)
    plot_data_from_file(file)
    
```
